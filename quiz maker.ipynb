```python
from tkinter import *
from tkinter.ttk import *


window = Tk()

window.title("Quiz")
window.geometry('500x500')


    
label1=Label(window,text='1.What year was the United Nations established?')

label1.pack()

selected = StringVar()

rad1 = Radiobutton(window,text='1944', value='1944', variable=selected)

rad2 = Radiobutton(window,text='1954', value='1954', variable=selected)

rad3 = Radiobutton(window,text='1945', value='1945', variable=selected)

rad1.pack()
rad2.pack()
rad3.pack()


label2=Label(window,text='2.How many elements are in the periodic table?')
label2.pack()

selected1 = StringVar()

rad4 = Radiobutton(window,text='118', value='118', variable=selected1)

rad5 = Radiobutton(window,text='119', value='119', variable=selected1)
rad6 = Radiobutton(window,text='117', value='117', variable=selected1)

rad4.pack()
rad5.pack()
rad6.pack()

label3=Label(window,text='3.Aureolin is a shade of what color?')

label3.pack()

selected2 = StringVar()

rad7 = Radiobutton(window,text='yellow', value='yellow', variable=selected2)

rad8 = Radiobutton(window,text='green', value='green', variable=selected2)
rad9 = Radiobutton(window,text='brown', value='brown', variable=selected2)

rad7.pack()
rad8.pack()
rad9.pack()


label4=Label(window,text='4. How many bones do we have in an ear?')

label4.pack()

selected3 = StringVar()

rad10 = Radiobutton(window,text='4', value='4', variable=selected3)

rad11 = Radiobutton(window,text='3', value='3', variable=selected3)
rad12 = Radiobutton(window,text='5', value='5', variable=selected3)

rad10.pack()
rad11.pack()
rad12.pack()


label5=Label(window,text='5. How many hearts does an octopus have?')

label5.pack()

selected4 = StringVar()

rad13 = Radiobutton(window,text='3', value='3', variable=selected4)

rad14 = Radiobutton(window,text='4', value='4', variable=selected4)
rad15 = Radiobutton(window,text='5', value='5', variable=selected4)

rad13.pack()
rad14.pack()
rad15.pack()

dic={1:'1945',2:'118',3:'yellow',4:'3',5:'3'}


def score():
    count=0
    if dic[1]==selected.get():
        count+=1
    if dic[2]==selected1.get():
        count+=1
    if dic[3]==selected2.get():
        count+=1
    if dic[4]==selected3.get():
        count+=1
    if dic[5]==selected4.get():
        count+=1
    entry.insert(END,count)
    btn.config(state='disabled')


btn = Button(window, text="Total Score",command=score)
btn.pack()


entry=Entry(window,width=5)
entry.pack()  

window.mainloop()
```
